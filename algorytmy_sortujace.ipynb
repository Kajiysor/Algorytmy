## Zad 1. Sortowanie bąbelkowe
Napisz funkcję `sort_bubble` sortującą rosnąco liczby zapisane w liście będącej parametrem formalnym funkcji. Pomyśl jaki kod byłby optymalny (czy obie pętle muszą przebiegać przez całą listę liczb). Zaprogramuj wersję optymalną. 

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
"""
Notacja o(n^2) - notacja kwadratowa, czas trwania jest zalezny od n czyli od dlugosci listy.

Pesymistyczna : o(n^2)
Optymistyczna : o(n^2)
Asymptotyczna : o(n^2)
"""


import random
Lista = [random.randint(-100,100) for x in range(25)]
print(Lista)
string = ['p','k','j','h','b','v','s','a','w','d']
print(string)

def bubble_sort(lista):
    """
    lista = lista wartosci do posortowania
    
    Funkcja zwraca podaną liste posortowaną rosnąco zgodnie z metodą Bubble Sort.
    """
    for i in range(len(lista)):
        for j in range(len(lista)-1-i):
            if lista[j] > lista[j+1]:
                temp = lista[j]
                lista[j] = lista[j+1]
                lista[j+1] = temp
    return lista

print(bubble_sort(Lista))
print(bubble_sort(string))
```

## Zad 2. Sortowanie przez wstawianie.
Napisz funkcję realizującą sortowanie przez wstawianie.

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
"""
Notacja o(n^2) - notacja kwadratowa, czas trwania jest zalezny od n czyli od dlugosci listy.
Pesymistyczna : o(n^2)
Optymistyczna : o(n)
Asymptotyczna : o(n^2)
"""
import random
Lista = [random.randint(-100,100) for x in range(25)]
print(Lista)
string = ['p','k','j','h','b','v','s','a','w','d']
print(string)

def wstawianie(lista):
    """
    lista = Lista wartosci do posortowania
    
    Funkcja zwraca podaną liste posortowaną rosnąco zgodnie z metodą Insertion Sort.
    
    """
    i = 1
    while i < len(lista):
        key = lista[i]
        j = i - 1
        
        while j >= 0 and key < lista[j]:
            lista[j + 1] = lista[j]
            j -= 1
        lista[j + 1] = key
        i += 1
    return lista

print(wstawianie(Lista))
print(wstawianie(string))
```

## Zad 3. Sortowanie przez scalanie
Napisz funkcję realizującą sortowanie przez scalanie.

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
"""
Notacja o(n log(n)) - notacja liniowo-logarytmiczna, czas trwania jest zalezny od n czyli od dlugosci listy.
Algorytm szybszy od tych z notacja kwadratowa.
Pesymistyczna : o(nlog(n))
Optymistyczna : o(nlog(n))
Asymptotyczna : o(nlog(n))
"""
import random
Lista = [random.randint(-100,100) for x in range(25)]
print(Lista)
string = ['p','k','j','h','b','v','s','a','w','d']
print(string)

def merge(L, R):
    """R i L: posortowane listy.
     Zwraca nowa posortowana niemalejaco liste o elementach z L i R."""
    W = []
    while L and R:
        W.append(L.pop(0) if L[0] < R[0] else R.pop(0))
    W.extend(R if R else L)
    return W


def merge_sort(T):
    """zwraca posortowana liste T"""
    if len(T) < 2: # baza rekurencji
        return T[:]
    else:  # część rekurencyjna
        middle = len(T) // 2
        L = merge_sort(T[:middle])
        R = merge_sort(T[middle:])
        return merge(L, R)

print(merge_sort(Lista))
print(merge_sort(string))
```

## Zad 4. Sortowanie przez zliczanie
Napisz funkcję realizującą sortowanie przez zliczanie.

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
import random
def counting_sort(lista):
    """ 
    Pesymistyczna wersja O(n)
    Optymistyczna wersja O(n)
    Srednio posortowana lista O(n)
    
    Algorytm przyjmuje nieposortowana liste, zwraca liste
    posortowana rosnaco
    """
    wynik = [0] * len(lista)
    najmniejszy = min(lista)
    najwiekszy = max(lista)
    zlicz = [0 for i in range(najmniejszy, najwiekszy + 1)]
    
    for i in lista:
        zlicz[i - najmniejszy] += 1
    
    for j in range(1, len(zlicz)): 
        zlicz[j] += zlicz[j - 1]
    
    for k in lista[::-1]:
        wynik[zlicz[k - najmniejszy] - 1] = k
        zlicz[k - najmniejszy] -= 1
    
    return wynik 
    
    

lista = counting_sort([random.randint(-10,10) for x in range(10)])
lista2 = counting_sort([random.randint(-50,10) for x in range(10)])
print(lista)
print(lista2)
```

## Zad 5. Sortowanie głupie
Jest to przykład złego algorytmu sortującego, które jednak daje zawsze poprawne wyniki. Zasada działania jest bardzo prosta. Przeglądamy kolejne pary sąsiednich elementów sortowanego zbioru. Jeśli bieżąco przeglądana para elementów jest w złej kolejności, elementy pary zamieniamy miejscami i całą operację rozpoczynamy od początku zbioru. Jeśli przeglądniemy wszystkie pary, zbiór będzie posortowany.

Jeżeli zbiór danych `d` ma długość `n`, to lista kroków jest nastepująca

    K01:	Dla i = 0, 1, 2, ..., n - 1: wykonuj K02...K04
    K02:	    Jeśli d[i] ≤ d[i + 1], to wykonaj następny obieg K01
    K03:	    Zamień miejscami d[i] oraz d[i + 1]
    K04:	    Idź do K01
    K05:	Zakończ

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
"""
Zlosonosc :
Pesymistyczna - Przy liscie nieuporządkowanej złożoność to O(n^3)
Asymptotyczna. - O(n^3)
Optymistyczna  - mozemy trafic na listę w której wystarczy jeden obrot pętli do posortowania, wtedy zlozonosc
to O(n)
"""
import random
Lista = [random.randint(-100,100) for x in range(25)]
print(Lista)
string = ['p','k','j','h','b','v','s','a','w','d']
print(string)

def dumb_sort(lista):
    """
    Funkcja jako argument przyjmuje listę wartości do posortowania
    
    Zwraca posortowaną rosnąco listę 
    """
    i = 0
    while i < len(lista) - 1:
        if lista[i] > lista[i + 1]:
            temp = lista[i + 1]
            lista[i + 1] = lista[i]
            lista[i] = temp
            i = 0
        else:
            i += 1
    return lista

print(dumb_sort(Lista))
print(dumb_sort(string))
```

## Zad 6. Bogosort
Napisz funkcję realizującą [bogosort](https://pl.wikipedia.org/wiki/Bogosort).

Jaką pesymistyczną / optymistyczną asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
"""
Zlozonosc pesymistyczna o(infinity), szczerze mozemy sie nigdy nie doczekac na to ze ten algorytm posortuje
nam daną liste. Mozemy sie spotkac ze algorytm zadziala bardzo szybko ale jest tez szansa na to aby nigdy
nie uzyskac rezultatu. Dzialanie algorytmu jest strasznie losowe, stad tez powyzsze wnioski.

Pesymistyczna : o(infinity)
Optymistyczna : o(1)
Asymptotyczna : o(n!)

"""
import random
Lista = [random.randint(-50,50) for x in range(5)]
string = ['u','t','h','f']
print(Lista)
print(string)
def sprawdzenie(lista):

    dlugosc = len(lista)
    for x in range(dlugosc - 1):
        if lista[x] >= lista[x + 1]:
            return False
    return True

def bogosort(lista):
    while (sprawdzenie(lista) == False):
        random.shuffle(lista)
        
    return lista


print(bogosort(Lista))
print(bogosort(string))
    
```

## Zad 7. Quicksort [40 pkt]
Zaimplementuj algorytm sortowania szybkiego. Funkcja powinna 

1. posiadać opcję ustawiania punktu podziału (pivot point) na (a) początku, (b) końcu, (c) w środku lub w (d) losowym miejscu listy
2. umożliwiać **opcjonalne** sortowanie krótkich fragmentów innymi metodami 
3. napisz program w wersjach: rekurencyjnej i iteracyjnej

Jaką pesymistyczną / optymistyczną / **średnią** asymptotyczną złożoność obliczeniową ma ten algorytm.

```python
"""
Zlozonosc quicksorta wynosi O(n log(n)). Sama nazwa wskazuje na jego szybkie dzialanie. Jest szybszy od tych
ze zlozonoscia np kwadratowa czy jeszcze gorsza.

Pesymistyczna : o(n^2)
Optymistyczna : o(n log(n))
Asymptotyczna : o(n log(n))
"""
import random
Lista = [random.randint(-100,100) for x in range(25)]
print(Lista)
string = ['p','k','j','h','b','v','s','a','w','d']
print(string)

def quicksortA(lista):
    """
    Funkcja quicksortA jako argument przyjmuje liste wartosci do posortowania wedlug jej pierwszego elementu
    
    Zwraca posortowana rosnąco liste.
    """
    mniejsze = [x for x in lista[1:] if x < lista[0]]
    wieksze = [x for x in lista[1:] if x > lista[0]]
    rowne = [x for x in lista if x == lista[0]]
    
    if len(lista) > 1:
        return quicksortA(mniejsze) + rowne + quicksortA(wieksze)
    else:
        return lista
    
def quicksortB(lista):
    """
    Funkcja quicksortB jako argument przyjmuje liste wartosci do posortowania wedlug jej ostatniego elementu
    
    Zwraca posortowana rosnąco liste.
    """
    
    mniejsze = [x for x in lista if x < lista[-1]]
    wieksze = [x for x in lista if x > lista[-1]]
    rowne = [x for x in lista if x == lista[-1]]
    
    if len(lista) > 1:
        return quicksortB(mniejsze) + rowne + quicksortB(wieksze)
    else:
        return lista
    
def quicksortRANDOM(lista):
    """
    Funkcja quicksortA jako argument przyjmuje liste wartosci do posortowania wedlug jej losowego elementu
    
    Zwraca posortowana rosnąco liste.
    """
    
    import random
    y = random.choice(lista)
    
    mniejsze = [x for x in lista if x < y]
    wieksze = [x for x in lista if x > y]
    rowne = [x for x in lista if x == y]
    
    if len(lista) > 1:
        return quicksortA(mniejsze) + rowne + quicksortA(wieksze)
    else:
        return lista
    
```

```python
# quciksort iteracyjnie
def quicksortIT(lista):
    """
    Funkcja sortuje dana liste przy uzyciu algorytmu QuickSort zaimplementowanego iteracyjnie
    
    zwraca posortowaną rosnąco listę
    """
    if len(lista) < 2:
        return lista
    final = []
    final.append(len(lista)-1)
    final.append(0)
    
    while final:
        l = final.pop()
        r = final.pop()
        
        idx = podzial(lista, l, r)
        
        if l < idx - 1:
            final.append(idx - 1)
            final.append(l)
        if r > idx + 1:
            final.append(r)
            final.append(idx + 1)
    return lista

def podzial(lista, start, end):
    pivot = lista[start]
    
    while start < end:
        while start < end and lista[end] >= pivot:
            end -= 1
        lista[start] = lista[end]
        
        while start < end and lista[start] <= pivot:
            start += 1
        lista[end] = lista[start]
        
    lista[start] = pivot
    return start

import random
Lista = [random.randint(-100,100) for x in range(25)]
print(Lista)
string = ['p','k','j','h','b','v','s','a','w','d']
print(string)

print(quicksortIT(Lista))
print(quicksortIT(string))
```
